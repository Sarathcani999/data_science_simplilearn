## Optimization function with one variable

Define function <i><b> f(x) = x<sup>2</sup> + 10sin(x) </b> </i>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optimize

%matplotlib inline
```

```python
f = lambda x : x**2 + 10*np.sin(x)
```

```python
f(3.1428/2)
```

```python
x = np.linspace( -10 , 10 , 500 )
```

```python
y = f(x)
```

```python
plt.plot(x , y , 'go--' , markersize = 1 )
plt.show()
```

```python
x_in = 10
x_min = optimize.minimize(f , x_in , method='BFGS').x[0]
y_min = f(x_min)
text = str(x_min) + ' , ' + str(y_min)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x , y , label = 'f(x) = x**2 + 10*sin(x)')
plt.plot(x_min , y_min , 'ro' , label = text )
#plt.plot(x_in, f(x_in) , 'go' , label = 'initial value')
plt.legend()
plt.show()
```

<b> Minimizing with constraints<b>

Minimise function 
<br>
V = l x b x h
<br>
<br>
Constraints :
<br>
l = 2 x b
<br>
l + b + h = 60

```python
def vol(x) :
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    
    return x1*x2*x3
```

To define constraints with inequality we have to define them as follows : <br><br>
<b>f(x) >= 0</b>


```python
# here there are no inequality , so
def con1_function(x) :
    return x[0] - 2*x[1]

def con2_function(x) :
    return x[0] + x[1] + x[2] - 60

con1 = {'type' : 'eq' , 'fun' : con1_function}
con2 = {'type' : 'eq' , 'fun' : con2_function}

cons = [con1, con2]
x0 = [100,100,100]
```

```python
b = [0.0 , 100.0]
optimize.minimize(vol , x0 , method='SLSQP' , constraints=cons , bounds=[b,b,b])
```
